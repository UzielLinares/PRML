```julia
include("../src/linear_regression.jl")
```

```julia
X = reshape(range(0, 10, length=100), :, 1)
y = [3 * x_ + randn() + 2 for x_ in X];
```

```julia
Φ = hcat(ones(length(X), 1), X);
```

```julia
lm = LinearRegression()
```

```julia
fit!(lm, Φ, y)
```

```julia
scatter(X, y, label="")
plot!(X, [(lm.coefs' * Φ[i, :])[1] for i in 1:size(Φ)[1]], ribbon=lm.σ * ones(size(Φ)[1]), label="Fit")
```
